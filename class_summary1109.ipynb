```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

```python
amp=1 #range[0.0, 1.0]
sr=10000 #sampling rate, Hz
dur=0.5 #in seconds
freq=100.0 #sine frequency, Hz
```

```python
theta=np.arange(0, 2*np.pi*5, 0.1) #*5는 다섯 바퀴 상징
```

```python
theta
```

```python
s=np.sin(theta)
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(theta, s, '.')
ax.set_xlabel('theta in radians')
ax.set_ylabel('value')
plt.show()
```

```python
t=np.arange(1, sr+1) #10000개 생성
t=np.arange(1, sr*dur+1)/sr #이렇게 하면 타임 완성 5000개 생성한 걸 10000으로 나눠서 시간 완성. 0.0001, 0.0002, ......0.5000 완성.
```

```python
t
```

```python
theta=t*2*np.pi*freq #np.pi는 파이를 지칭하는 거임. 타임에 2파이를 곱하면? 0초에서 1초에 2파이를 곱하면? =0파이~2파이=그냥 한 바퀴
#따라서 freq을 곱해줌
```

```python
s=np.sin(theta) #signal 생성, 사인 이용
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000], '.') #이 그래프서 점의 개수는? 1000 왜 1000개로 개수를 지정하냐면 개수 지정을 안하면 거의 선으로 나옴
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
```

```python
c=np.exp(theta*1j)
c
```

```python
c.real
```

```python
fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.plot(t[0:100], c.real[0:100], c.imag[0:100], '.') #real은 a+bi에서 a만(실수_real num만) 가져오는 역할. 
ax.set_xlabel('time(s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
plt.show()
```
